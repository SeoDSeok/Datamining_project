### 한꺼번에 수행하고자 get_dummies 라는 함수 사용

```python
dummy = pd.get_dummies(data, columns = ['지점','발생해역', '기상상태', '선 종'])
```

### 그렇게 생성한 data와 dummy들을 합치기

```python
df = pd.concat([data,dummy],axis=1)
df
```

```python
df.drop(['지점', '발생해역', '선 종', '기상상태'], axis=1, inplace=True)
```
