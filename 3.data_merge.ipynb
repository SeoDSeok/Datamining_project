### weather데이터 2020이후 데이터 추가로 적용

```python
data3 = pd.read_csv('weather_final2.csv', encoding='cp949')
```

```python
data3
```

```python
data3 = data3.replace('거문도', '여수')
data3 = data3.replace('거제도', '부산')
data3 = data3.replace('울릉도', '속초')
data3 = data3.replace('칠발도', '목포')
data3 = data3.replace('추자도', '제주')
data3 = data3.replace('외연도', '보령')
data3 = data3.replace('서해206', '평택')
data3 = data3.replace('서해170', '태안')
data3 = data3.replace('신안', '완도')
data3['지점'].unique()
```

```python
data3 = data3[data3.지점 != '마라도']
data3 = data3[data3.지점 != '덕적도']
data3 = data3[data3.지점 != '가거도']
data3 = data3[data3.지점 != '홍도']
data3
```

```python
time = []
for str in data3['일시']:
    if (not(type(str) == float)):
        tap = str[11:13]
        if ':' in tap:
            tap = tap[0]
            time.append(tap)
        else:
            time.append(tap)
len(time)
```

```python
data3['시간대별'] = time
```

```python
day = []
for str in data3['일시']:
    day.append(str[0:10])
len(day)
```

```python
data3['key'] = day
```

```python
data3 = data3.astype({'시간대별':'int'})
data3.info()
```

### weather data 2011 ~ 2021생성하기위해 concat

```python
df = pd.concat([data2, data3], axis = 0)
df
```

### 최종적인 데이터 생성

```python
final3 = pd.merge(data, df, on=['key','시간대별', '지점'])
final3
```
